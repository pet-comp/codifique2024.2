## O que são estruturas de repetição?

São instruções utilizadas para repetir um determinado bloco de código até que seja atingida uma condição de parada.
Tais estruturas também são chamdas de **laços** ou **loops**.

### Exemplo inicial:

Suponha que você precise de um prorama para ler 3 números inteiros.

````python
# Solução para ler 3 inteiros

print("Informe um numero inteiro: ")
n1 = int(input())
print("Informe um numero inteiro: ")
n2 = int(input())
print("Informe um numero inteiro: ")
n3 = int(input())

print("Os numeros digitados foram {}, {}, {}".format(n1, n2, n3))
````

## Estruturas de repetição da linguagem Python

E se quisessemos ler 10, 100, 1000 inteiros?

Certamente seria inviável de fazer apenas com *inputs*. Essa é uma das funções das estruturas de repetição.

Em Python, existem dois tipos de laços de repetição: O **while** e o **for**.

* While: Funciona de forma muito parecida com uma **condicional**, porém aqui o código se repete até que a condição se mostre falsa.

* For: Muito utilizada para percorrer intervalos, conjuntos, listas, em que já temos noção do critério de parada.

Podemos também adicionar o comando **else** para que o programa realize uma instrução ao fim de um **loop**.

<br> Exemplos: Imprimir os números de 0 a 10 utilizando as estruturas citadas acima.

````python
n = 0

while(n <= 10):
  print("{} ".format(n))
  n += 1
````

````python
for i in range(0, 11):
  print("{}".format(i))
````

### Resposta para a pergunta:

Como faríamos para ler uma quantidade indeterminada de números inteiros?

## O operador **range()**:

Este operador é de grande ajuda para lidarmos com loops, principalmente loops em **for**. Basicamente, geramos o intervalo definido entre dois números, o qual pode ser percorrido por uma variável dentro de um laço de repetição.

````python
# Solução generalizada, ler n inteiros (com "for")

print("Informe o numero de inteiros:")
quantidade = int(input())

for i in range (0, quantidade):
  n = int(input())
  print("{}° numero: {}".format((i + 1), n))

else: print("Fim dos numeros")
````

````python
# Solução generalizada, ler n inteiros (com "while")

print("Informe o numero de inteiros:")
quantidade = int(input())
contador = 0

while(contador < quantidade):
  num = int(input())
  print("{}° numero: {}".format((contador + 1), num))
  contador += 1

else: print("Fim dos numeros")
````

## Percorrendo **strings** com estruturas de repetição:

Podemos utilizar **loops** para percorrer uma **string**, de modo que acessamos um de seus caracteres de maneira sequencial a cada iteração do laço.


````python
# Percorrendo uma "string" com um laço

string = input("Digite uma palavra: ")
count = 1

print("A palavra eh: {}".format(string))

for i in string:
  print("{}ª letra: {}\n".format(count, i))
  count += 1
````

## Estruturas **predeterminadas** X **não predeterminadas**

As estruturas de repetição podem ser tanto predeterminadas quanto indeterminadas.
Uma estrutura **predeterminadas (1)** é aquela em que sabemos, antes de sua execução, o momento em que as iterações serão interrompidas, ou seja, o momento em que seu critério de parada será verdadeiro.

Já as estruturas **não predeterminadas (2)** são iniciadas sem que tenhamos noção de quando seu critério de parada será verdadeiro, sendo essa condição muitas vezes definida durante a realizadação do bloco de código dentro do loop.

* Exemplo 1: Imprimir números inteiros em um dado intervalo.

* Exemplo 2: Realizar um laço que peça ao usuário um número par e não seja encerrado enquanto o número par não for digitado.

````python
# Exemplo de estrutura pré-determinada:
# Imprimindo todos os numeros de um dado intervalo

# Lendo o inicio e fim do intervalo pedido pelo usuário
print("Insira o primeiro numero do intervalo: ")
primeiro = int(input())
print("Insira o ultimo numero de intervalo: ")
ultimo = int(input())

# Imprimindo o intervalo
for i in range(primeiro, ultimo + 1):
  print("{}".format(i), end=" ")

else: print("\nFim do intervalo")
````

````python
# Exemplo de estrutura NÃO pré-determinada:
# Pedindo ao usuário para digitar um numero par

print("Informe um numero par:")
n = int(input())

# Enquanto número digitado não for par
while((n % 2) != 0):

  # Exibindo uma mensagem de erro e solicitando novamente o número
  print("Input invalido, informe um numero par!")
  n = int(input())

else: print("Fim do laço")
````

## Quando utilizar **for** ou **while**?

Como podemos perceber, o **for** é ideal para percorrer intervalos e sequências, no entanto, não lida tão bem para casos em que o critério de parada é desconhecido previamente.

Assim, de modo geral, tendemos a utilizar o **for** para estruturas **predeterminadas** (quando conhecemos de antemão o momento em que o laço será encerrado), e o **while** para estruturas **não predeterminadas**.

## Operadores auxiliares: **Break** e **Continue**

Para trabalhar com estruturas de repetição, por vezes é necessário utilizar alguns operadores além dos que já conhecemos até o momento. Entre eles, estão o **break** e **continue**, explicados abaixo:

### Break:
Este operador nos permite interromper a execução de um laço sem verificar seu critério de parada. Geralmente é utilizado junto à uma **condicional** que determina quando devemos interromper o laço.

### Continue:

O **continue** tem a função de ignorar todo o bloco de código abaixo dele. Ou seja, diferentemente do **break** (que encerra o laço), o **continue** simplesmente "pula" para a próxima iteração, não realizando qualquer código que haja após ele.


````python
# Interrompendo um laço de repetição utilizando o "break"

print("Em qual numero o laço deve ser interrompido?")
n = int(input())

for i in range(0, 100):
  print("{} ".format(i))

  # Caso "i" seja igual ao número informado pelo usuário,
  # encerramos o laço através do "break"
  if(n == i):
    print("Laço interrompido!")
    break
````

````python
# Ignorando um bloco de código com o "continue"

n_impressos = 0

print("Digite 'S' para imprimir o numero")

for i in range(0, 6):
  op = input()

  # Caso o usuário não tenha digitado 'S', o programa ignora
  # o restante do bloco de código
  if(op != 'S'):
    continue

  print("{}".format(i))
  n_impressos += 1

print("\nForam impressos {} numeros".format(n_impressos))
````

## Laços aninhados: Utilizando um Loop dentro de outro

Em diversas situações na computação precisaremos iterar uma segunda variável dentro de um laço de repetição já existente. Para isso, utilizamos **laços aninhados**, os quais não são nada além de um laço executado dentro de outro.

<br> Exemplo: Imprimir a tabuada dos números de 1 a 9

````python
# Imprimindo as tabuadas do 1 ao 9

# O primeiro laço (for) percorre todas as tabuadas que queremos imprimir

for n1 in range(1, 10):

  print("Tabuada do {}:\n".format(n1))

  # O segundo (while) imprime a tabuada
  # da variável inicializada no primeiro laço

  n2 = 1
  while(n2 <= 10):
    print("{} X {} = {}".format(n1, n2, n1*n2))
    n2 += 1
  else: print("\n")
````
