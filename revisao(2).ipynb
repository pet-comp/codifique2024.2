# Entradas e saídas

### Entrada

A entrada de dados em Python é feita com a função **input()**, que permite ao usuário digitar algo no teclado e o programa armazenar essa informação para usar depois. É como se o computador perguntasse algo e esperasse sua resposta!


##### Exemplo: 

##### Receber um nome:

```python
nome = input()
```
*Aqui, o que for digitado será armazenado na variável nome.*


##### Receber um valor com uma mensagem personalizada:
```python
salario = input("Digite o seu salário:")
```
*O que vai aparecer na tela ao rodar o código:*

<span style='color: #90EE90'> Digite o seu salário:</span>

*O programa só continua depois que você digitar algo e apertar Enter!*
<br>
<br>


No último exemplo, ao colocar alguma mensagem dentro dos parânteses, a mensagem aparecerá primeiro e depois será permitido que você digite!

**Ao rodar a linha de código 2, o seguinte aparecerá na tela:**

##### Digite o seu salário: 

O código só continuará depois que alguma coisa seja digitada

### Casting (Conversão de Tipos)

Às vezes, o que digitamos com o *input()* precisa ser convertido para outro tipo de dado. Isso é feito com casting, ou seja, a conversão de um tipo de dado para outro.

##### Exemplo 1 - Transformar um texto em número:

Quando usamos **input()**, o valor digitado é sempre tratado como texto (string). Se quisermos trabalhar com números (como somar, multiplicar etc.), precisamos converter para **inteiro** ou **decimal**

```python
idade = int(input("Digite a sua idade: "))

print(idade + 1)  # Soma 1 na idade
```
*O que vai aparecer na tela:*

<div style='color: #90EE90'> 

Digite a sua idade: 25

26

</div>



##### Exemplo 2 - Transformar um número em texto:

Às vezes, queremos juntar um número com uma frase. Para isso, precisamos converter o número para texto.

```python
idade = 25

print("Você tem " + str(idade) + " anos.")
```
*O que vai aparecer na tela:*


<span style='color: #90EE90'> Você tem 25 anos.</span>


Aqui, usamos str() para transformar o número 25 em texto para que ele possa ser combinado com a frase.

# Variáveis 
## **O que são Variáveis?**

As variáveis são **"caixinhas"** onde guardamos valores que podemos usar ao longo do nosso código. Elas nos permitem armazenar diferentes tipos de dados, como números, textos e até resultados de operações.

### Como funciona? 

- Você dá um nome para a caixinha (variável).
- Coloca algo dentro dela (um valor).
- E depois pode usar esse valor sempre que precisar!

### Exemplo:
```python
nome = "Ana"
idade = 20
``` 
Aqui, criamos duas variáveis:

- nome guarda o valor "Ana".
- idade guarda o valor 20.

### Regras para Nomes de Variáveis

1. O nome da variável não pode começar com um número.
2. Não use espaços, substitua-os por _ (underline).
3. Não pode usar palavras reservadas do Python (como print, input).
4. Use nomes que façam sentido para o que você está guardando.


### Exemplos válidos:
```python
meu_nome = "Carlos"
salario_atual = 1500.50
```
### Exemplos inválidos:
```python
2nomes = "Maria e João"  # Não pode começar com número
meu nome = "Carlos"      # Não pode ter espaço
```

### Mudando o Valor de uma Variável

O valor de uma variável pode ser alterado ao longo do código, basta reatribuir um novo valor a ela.

#### Exemplo:

## Tipos de Variáveis

Em Python, você pode guardar diferentes tipos de valores em variáveis. Aqui estão alguns dos tipos mais comuns:

1. **Inteiros (int):** Números inteiros, como `1`, `2`, `100`.
    ```python
    idade = 25
    ```

2. **Decimais (float):** Números com ponto decimal, como `1.5`, `3.14`.
    ```python
    altura = 1.75
    ```

3. **Textos (strings):** Sequências de caracteres, como `"Olá"`, `"Python é legal!"`.
    ```python
    nome = "Carlos"
    ```

4. **Booleanos (bool):** Verdadeiro ou falso, como `True`, `False`.
    ```python
    esta_chovendo = False
    ```

### Exemplo de uso:
```python
nome = "João"
idade = 21
altura = 1.80
esta_aprovado = True

print(nome, idade, altura, esta_aprovado)
```

*O que vai aparecer na tela:*

João 21 1.8 True

# Listas

Uma lista é como uma **coleção de itens** (como uma fila de coisas), onde podemos armazenar vários valores em uma única variável. Esses valores podem ser de qualquer tipo: números, textos, ou até outras listas!

## Criando uma lista

Para criar uma lista, colocamos os itens entre colchetes [ ] e separamos por vírgulas.

### Exemplo 1: Criando uma lista com números

```python
numeros = [10, 20, 30, 40]
print(numeros)
```

*O que vai aparecer na tela:*
<div style='color: #90EE90'> 
[10, 20, 30, 40]
</div>

Aqui, a lista chamada numeros tem quatro elementos: 10, 20, 30 e 40.

## Acessando Itens da Lista

Cada item em uma lista tem uma posição (chamada de índice) que começa no número 0.

### Exemplo: Acessar o primeiro item de uma lista

```python
nomes = ["Ana", "Carlos", "Maria"]
print(nomes[0])  # Primeiro item
```

*O que vai aparecer na tela:*
<div style='color: #90EE90'> 
Ana
</div>


## Modificando Itens da Lista

Podemos mudar o valor de um item na lista indicando a sua posição (índice).

### Exemplo: Alterar o segundo nome da lista

```python
nomes = ["Ana", "Carlos", "Maria"]
nomes[1] = "João"   # Altera o segundo item
print(nomes)
```
*O que vai aparecer na tela:*
<div style='color: #90EE90'> 
['Ana', 'João', 'Maria']
</div>

# Condicionais

As condicionais são usadas para tomar decisões no seu código. Elas permitem que o programa execute diferentes ações com base em condições específicas. Em Python, usamos a palavra-chave `if` para começar uma condição.


## Estrutura Básica

A estrutura básica de uma condição é a seguinte:


```python
if condição:
    # código a ser executado se a condição for verdadeira
```

**<span style = "color: red">IMPORTANTE:</span> Para que o computador entenda o que está dentro do if e o que não está, é preciso <span style = "color: orange">identar</span> seu código corretamente!**
Ou seja:

```python
if condição:
    # código dependente da condição. Será executado se a condição for verdadeira   
    # TODAS as linhas que se alinham dessa forma estarão dentro do BLOCO if 
# código fora do if! Será executado independente se a condição for verdadeira ou não
```

### Exemplo 1: Verificar idade


Vamos verificar se uma pessoa pode votar:

```python
idade = int(input("Digite sua idade: "))

if idade >= 18:
    print("Você pode votar!")
```

#### O que acontece aqui:

- O programa pede a idade do usuário.
- Se a idade for 18 ou mais, imprime "Você pode votar!".

## Condicionais com `else`

Podemos adicionar uma condição alternativa usando `else`. Isso define o que deve acontecer se a condição não for verdadeira.

### Exemplo 2: Verificar idade com else
```python
idade = int(input("Digite sua idade: "))

if idade >= 18:
    print("Você pode votar!")
else:
    print("Você não pode votar ainda.")
```
O que vai acontecer:

- Se a idade for 18 ou mais, o programa diz que a pessoa pode votar.
- Se a idade for menor que 18, diz que não pode votar.


## Condicionais com `elif`

Se quisermos verificar mais de uma condição, usamos `elif` (que significa "else if"). Isso permite testar várias condições diferentes.


### Exemplo 3: Verificar faixa etária

```python
idade = int(input("Digite sua idade: "))

if idade < 13:
    print("Você é uma criança.")
elif idade < 18:
    print("Você é um adolescente.")
else:
    print("Você é um adulto.")
```

O que vai acontecer:

- Se a idade for menor que 13, imprime "Você é uma criança."
- Se a idade for entre 13 e 17, imprime "Você é um adolescente."
- Se a idade for 18 ou mais, imprime "Você é um adulto."

***

## Comparações nas Condicionais

Nas condicionais, comparamos valores usando **operadores de comparação**. Eles são usados para verificar se uma condição é verdadeira ou falsa. Vamos ver os principais operadores:

###  1. `AND` (e) - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -

O operador `and` é usado quando queremos que **todas** as condições sejam verdadeiras ao mesmo tempo.

#### Exemplo - Já pode sair de casa?:
```python
dinheiro = input("Você tem dinheiro? (sim/não): ")
dever_casa = input("Você já fez o dever de casa? (sim/não): ")

if dinheiro == "sim" and dever_casa == "sim":
    print("Você pode sair com seus amigos!")
else:
    print("Você não pode sair com seus amigos.")
```
O que acontece aqui:

- O programa pergunta se você tem dinheiro e se já fez o dever de casa.
- **Se** <span style ="color: #FFA500">ambas</span> as respostas forem "sim", imprime "Você pode sair com seus amigos!".
- **Se** qualquer uma das respostas for "não", imprime "Você não pode sair com seus amigos."

### 2. `OR` (ou)  - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -

O operador `or` é usado quando queremos que **pelo menos uma** das condições seja verdadeira.

```python
dia = input("Digite o dia da semana: ")

if dia == "sábado" or dia == "domingo":
    print("É fim de semana! Você pode descansar ou sair com os amigos.")
else:
    print("É dia de semana! Hora de estudar ou ir para a escola.")

```

O que acontece aqui:

- O programa pergunta o dia da semana.
- Se o dia for "sábado" ou "domingo", imprime "É fim de semana! Você pode descansar ou sair com os amigos."

### 3. IGUAL `==`  - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -

O operador == é usado para verificar se dois valores são iguais.

```python

seu_jogo_favorito = input("Qual é o seu jogo favorito? ")
meu_jogo_favorito = "Minecraft"

if seu_jogo_favorito == meu_jogo_favorito:
    print("Nós dois amamos Minecraft!")

```

O que acontece aqui:
- O programa pergunta qual é o jogo favorito da pessoa.
- Se a resposta for "Minecraft", assim como está inserido na variável `meu_jogo_favorito`,ele responde com "Nós dois amamos Minecraft!"


**<span style = "color: green">Nota:</span> Não confunda = (atribuição de valor) com == (comparação de igualdade)!**

#### Exemplos da diferença entre **=** e **==**

A seguir, estamos comparando se o conteúdo da variável `seu_jogo_favorito` é **igual** a `meu_jogo_favorito`

```python
seu_jogo_favorito == meu_jogo_favorito
```

O operador != é usado para verificar se dois valores são diferentes.

```python

bebida_favorita = input("Qual a sua bebida favorita? ")

if bebida_favorita != "refrigerante":
    print("Você gosta de algo diferente de refrigerante!")
else:
    print("Você gosta de refrigerante, assim como eu!")


```

O que acontece aqui:

- O programa pergunta qual é a bebida favorita da pessoa.
- Se a resposta não for "refrigerante", imprime "Você gosta de algo diferente de refrigerante!"
- Se a resposta for "refrigerante", imprime "Você gosta de refrigerante, assim como muitas pessoas!"

**<span style = "color: green">Nota:</span> o operador ! é um sinal de negação, é como se estivéssemos falando "não igual" no lugar de !=**

# Loops ou Laços


Looping é uma forma de repetir um bloco de código várias vezes. É útil quando queremos executar a mesma ação várias vezes sem precisar escrever o mesmo código várias vezes!


## Tipos de Loops

### 1. Loop `for`

O loop for é usado para repetir um bloco de código **um número específico de vezes**. Geralmente, ele é usado para **percorrer listas, strings ou intervalos de números**.

#### Estrutura 1 

```python
for i in range(x):
    #código que será executado x vezes, contando de 0 a x-1
```
**<span style = "color: green">Comentário:</span> Estamos definindo que a variável <span style ="color: #FF00FF">i</span> que vai contar de <span style="color: #FF00FF">0</span> até um número <span style="color: #FF00FF">x - 1</span>, como vemos a seguir, de forma mais definida:**

#### Exemplo da Estrutura 1
```python
for i in range(5):
    print(i)
```
*O que vai aparecer na tela:*
<div style='color: #90EE90'> 
0<br>
1<br>
2<br>
3<br>
4<br>
</div>

#### Explicação:

- O range(5) cria uma sequência de números de 0 a 4.
- O loop for percorre cada número e o imprime.

#### Estrutura 2:
```python
for i in range(x, y):
    # código que será executado quando i chegar em y-1, partindo de x
```
**<span style = "color: green">Comentário:</span> Aqui estamos definindo que a variável <span style ="color: #FF00FF">i</span> que vai contar de um número <span style="color: #FF00FF">x</span> até um número <span style="color: #FF00FF">y - 1</span>, como vemmos a seguir, de forma mais definida:**

#### Exemplo da Estrutura 2

```python
for i in range(3, 8):
    print(i)
```
*O que vai aparecer na tela:*
<div style='color: #90EE90'> 
3<br>
4<br>
5<br>
6<br>
7<br>
</div>

#### Estrutura 3:
```python
for i in range(x, y, z):
    # i vai contar de x até y-1, mas pulando de z em z
```

**<span style = "color: green">Comentário:</span> E finalmente, definindo que a variável <span style ="color: #FF00FF">i</span> que vai contar de um número <span style="color: #FF00FF">x</span> até um número <span style="color: #FF00FF">y - 1</span>, mas <span style="color: #FF00FF">pulando de z em z</style> como vemos a seguir, de forma mais definida:**

#### Exemplo da Estrutura 3

```python
for i in range(0, 30, 5):
    print(i)
```

*O que vai aparecer na tela:*
<div style='color: #90EE90'> 
0<br>
5<br>
10<br>
15<br>
20<br>
25<br>
</div>

### 2. Loop `while`

O loop while repete um bloco de código enquanto uma condição for verdadeira. É útil quando **não sabemos quantas vezes precisamos repetir o código**.

#### Estrutura:
```python
while condição:
    # código que executa ENQUANTO a condição é verdadeira
```

#### Exemplo:
```python
contador = 0  # chamamos a prática de colocar um valor inicial em uma variável de inicialização ou inicializar uma variável
while contador < 5:
    print(contador)
    contador += 1  # Adiciona 1 ao contador, é como se fosse contador = contador + 1
```

*O que vai aparecer na tela:*
<div style='color: #90EE90'> 
0<br>
1<br>
2<br>
3<br>
4<br>
</div>

#### Explicação:

- O loop continua enquanto contador for menor que 5.
- Dentro do loop, contador é impresso e, em seguida, incrementado em 1
